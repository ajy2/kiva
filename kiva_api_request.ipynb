```python
import requests
import json
import time

base_url = 'https://api.kivaws.org/v1/loans/'


for i in range(84, 1000):
    query = str(i)+'.json'
    r = requests.get(base_url+query)
    
    try:
        if(str(r)!='<Response [403]>'):
            print(str(i) + " : " + str(r))

            path= './kiva_dataset/loan/'+query
            with open(path,'w',encoding='utf-8') as make_file:
                json.dump(r.json(), make_file)
            time.sleep(1)
        else:
            print("taking a nap")
            time.sleep(1800)
            print(str(i) + " : " + str(r))

            path= './kiva_dataset/loan/'+query
            with open(path,'w',encoding='utf-8') as make_file:
                json.dump(r.json(), make_file)
            time.sleep(1)
    except:
        pass
```
